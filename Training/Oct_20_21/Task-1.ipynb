from www.google.co.in URL extract all the HYPER LINKS  

and store them in a CSV file  

HYPERLINK  and DESCRIPTION should be store in the CSV file


```python
import urllib.request
from bs4 import BeautifulSoup
```

```python
# opening url and extracting Html content using BS
html = urllib.request.urlopen('https://www.google.co.in')
soup = BeautifulSoup(html,'lxml')
# finding attribute 'a'
links_list = soup.find_all('a')

links = []
desc = []
# storing the description
for i in links_list:
    desc.append(i.text)
# storing the links  
for link in links_list:
    if 'href' in link.attrs:
        links.append(link.attrs['href'])

```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'Hyperlink':links,
             'Description': desc})
df
```

```python
df.to_csv('googlelinks.csv')
```
